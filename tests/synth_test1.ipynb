```python
from dlqpiler import parser
from dlqpiler import synth
from qiskit import Aer
from ply import yacc
```

```python
code = yacc.parse('''
    a1[1] in {0, 1};
    a2[1] in {0, 1};
    a3[1] in {0, 1};
    b[1] := (a1) and (a2) and (not a3);
    amplify b 1 times
''')
qe = synth.QuantumEvaluator(code)
qe.build_all()
qe.quantum_circuit.draw(output='mpl')
```

```python

qe.simulate(Aer.get_backend('aer_simulator'))
```
